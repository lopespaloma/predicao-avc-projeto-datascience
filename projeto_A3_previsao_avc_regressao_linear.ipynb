## TREINAMENTO MODELO - REGRESS√ÉO LINEAR


```python
import pandas as pd
import numpy as np
```

```python
#arquivo csv - https://drive.google.com/file/d/1nqO1CM37CHyMd9bi4jG81Myd-iL_mSyX/view?usp=sharing

avc_linear = pd.read_csv('dados-regressao-linear.csv', delimiter= ',')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
avc_linear.head(500)
```

```python
x = avc_linear[['Genero', 'Idade', 'Hipertensao', 'Doenca Cardiaca', 'Estado Civil',
        'Tipo de trabalho', 'Localizacao Residencial', 'Nivel medio de glicose',
        'IMC', 'Condicao de fumante']]
y = avc_linear['AVC']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=50)
```

```python
modelopreditor = LinearRegression()
modelopreditor.fit(X_train, y_train)
coeff = pd.DataFrame(modelopreditor.coef_, x.columns, columns = ['Coeficiente'])
coeff
```

```python
predictions = modelopreditor.predict(X_test)
X_test.head()
```

```python
print(predictions)
```

```python
y_test.head()
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
